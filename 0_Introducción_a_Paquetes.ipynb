```python
games_df.isnull().sum().plot(kind='bar')
```

Y esto a cuánto porcentaje del dataframe equivale?

```python
missing_percentage = games_df.isnull().sum().div(games_df.shape[0]) * 100
missing_percentage.plot(kind='bar')
#Cerca del 60%
```
